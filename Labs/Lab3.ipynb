```python
from matplotlib import pyplot as plt
import numpy as np
from math import sqrt, pi, e
from scipy.stats import norm

np.random.seed(2)
```

```python
def euclideanDistance(x1, x2):
  return np.sqrt((x1 - x2)**2)


def knn(k, r, X1):
  distances = {}
  for i in range(len(X1)):
    dis1 = -1
    dis2 = -1

    if i-k > 0:
      dis1 = euclideanDistance(X1[i], X1[i-k])

    if i+k < len(X1):
      dis2 = euclideanDistance(X1[i], X1[i+k])

    distances[i] = max(dis1, dis2)


  anomalies = sorted(distances.items(), key=lambda x:x[1], reverse=True)[:int(m*r)]
  anomalies = [x[0] for x in anomalies]
  anomaliesToPlot = [X1[i] for i in anomalies]
  return anomaliesToPlot
```

```python
m = 1000
m1 = int(0.6*m)
m2 = m - m1

X = np.concatenate([np.random.normal(-7, 1, m1), np.random.normal(-7, 1, m2)])
X1 = np.sort(X)
Y = 0.6*norm.pdf(X1,-7,1) + 0.4*norm.pdf(X1, -7, 1)

fig, ax = plt.subplots(6, 3, figsize=(18, 16))
k_set = [1, 2, 5, 10, 20, 50]
r_set = [0.01, 0.05, 0.1]
for i, k in enumerate(k_set):
  for j, r in enumerate(r_set):
    anomaliesToPlot = knn(k, r, X1)
    ax[i, j].set_title(f'k = {k}, r = {r}')
    ax[i, j].set_xlim(-10, 10)
    ax[i, j].set_ylim(0, 0.5)
    ax[i, j].plot(X1, Y)
    ax[i, j].scatter(anomaliesToPlot, [0]*len(anomaliesToPlot), color='red', s=150, marker='|')

plt.tight_layout()
plt.show()
```
