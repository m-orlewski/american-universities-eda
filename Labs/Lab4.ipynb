```python
from matplotlib import pyplot as plt
import numpy as np
from math import sqrt, pi, e
from scipy.stats import norm
from sklearn.cluster import AgglomerativeClustering

np.random.seed(2)
```

1. generalnie sredni daje najlepsze wyniki, ale nie jest odporny na nieliniowe przeksztalcenia danych, np. logarytm

2. dla 5 klastrow raczej wszystkie wyniki sa zle, wiemy ze rozklad jest 2/3 modalny

3. nie ma idealnego rozwiazania, starajmy sie dobrac te mniej zle

```python
# zad. 1 klasteryzacja
m = 1000
m1 = int(0.3*m)
m2 = int(0.4*m)
m3 = m - m1 - m2

means = [1, 1.5, 2, 2.5, 3]
linkages = ['ward', 'complete', 'average']
linkages_names = ['najblizsze', 'najdalsze', 'srednie']

fig, ax = plt.subplots(5, 3, figsize=(20, 20))
for i, mean in enumerate(means):
  X1 = np.random.normal(-6, 1, m1)
  X2 = np.random.normal(0, 1, m2)
  X3 = np.random.normal(mean, 1, m3)
  X = np.concatenate((X1, X2, X3))
  X = np.sort(X)
  Y = 0.3*norm.pdf(X,-6,1) + 0.4*norm.pdf(X, 0, 1) + 0.3*norm.pdf(X, mean, 1)
  for j, (linkage, linkage_name) in enumerate(zip(linkages, linkages_names)):
    clustering = AgglomerativeClustering(n_clusters=3, linkage=linkage).fit(X.reshape(-1, 1))
    colormap = np.array(['red', 'blue', 'green'])
    ax[i, j].set_title(f'mean={mean}, linkage={linkage_name}')
    ax[i, j].grid(True)
    ax[i, j].set_xlim(-10, 10)
    ax[i, j].set_ylim(0, 0.5)
    ax[i, j].plot(X, Y, c='gray')
    ax[i, j].scatter(X, Y, s=5, c=colormap[clustering.labels_])

plt.tight_layout()
plt.show()
```
